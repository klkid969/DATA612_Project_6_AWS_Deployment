## Amazon S3 – Long-Term Storage

The first AWS service used in this project is Amazon Simple Storage Service (Amazon S3). I selected S3 because the recommender system requires a reliable location to store the MovieLens dataset and any supporting project files.

Rather than storing the data only on my local computer, S3 provides cloud-based storage that can be accessed by other AWS services, making it a good foundation for a cloud-based recommendation system.

One way to think about Amazon S3 is as a secure digital storage room. Just as an office stores paper files in a file room so employees can retrieve them when needed, Amazon S3 stores datasets and project files so cloud applications can access them whenever they are needed.

### Why This Step Matters

Every cloud application needs a place to store its data. Amazon S3 serves as that storage location by keeping files available, durable, and accessible when needed. Choosing S3 also allows the recommender system to integrate with other AWS services without relying on files stored only on a personal computer.

### Uploading the MovieLens Dataset

After creating the Amazon S3 bucket, I uploaded the MovieLens 1M dataset, including the `movies.dat`, `ratings.dat`, and `users.dat` files. These files will serve as the data source for the Spark-based recommender system.

Storing the dataset in Amazon S3 provides a centralized location that can be accessed by other AWS services, such as Amazon EC2. This approach more closely reflects how data is managed in real-world cloud environments, where datasets are stored separately from the computing resources that process them.

### Why This Step Matters

Uploading the dataset to Amazon S3 separates data storage from data processing. Instead of keeping the files only on a personal computer, the recommender system can retrieve the data directly from cloud storage. This makes the solution easier to manage and allows the same dataset to be used by different cloud resources when needed.

## Amazon EC2 – Cloud Computing

The second AWS service used in this project is Amazon Elastic Compute Cloud (Amazon EC2). I selected EC2 because it provides a virtual computer in the cloud where applications can be installed and executed without relying on my personal computer.

For this project, EC2 represents the environment where the Spark-based recommender system could be deployed and executed while accessing the MovieLens dataset stored in Amazon S3.

### Why This Step Matters

Amazon S3 stores the data, but it does not perform any calculations. Amazon EC2 provides the computing power needed to process the dataset and run the recommender system.

Separating storage from computing is a common cloud design because each service has a specific responsibility. This approach also makes it easier to use a larger or more powerful virtual computer later if the recommender system grows.

### Launching an Amazon EC2 Instance

After storing the MovieLens dataset in Amazon S3, the next step is to create an Amazon EC2 instance. An EC2 instance is a virtual computer that can be used to run applications, process data, and execute Spark workloads in the cloud.

For this project, I selected a Free Tier eligible EC2 instance because it provides enough computing resources to demonstrate how a Spark recommender system could be deployed while avoiding unnecessary costs.

### Why This Step Matters

Unlike Amazon S3, which is responsible for storing files, Amazon EC2 provides the computing resources needed to process data. Separating storage from computing allows each service to perform a specific task and makes it easier to increase computing resources in the future if the recommender system grows.

Using a Free Tier eligible EC2 instance through my AWS trial account allowed me to learn how cloud computing works without purchasing additional hardware. Although the account includes trial benefits, it also introduced me to the same cloud services and deployment process that organizations use in larger production environments.

### Launching the Amazon EC2 Instance

After configuring the Amazon EC2 instance, I launched a Free Tier eligible virtual machine using Amazon Linux 2023. The instance was created successfully and reached a running state, confirming that the cloud computing environment was available for use.

Although this project did not require running the recommender system on the instance, the deployment demonstrated how Amazon EC2 can provide the computing resources needed to process data stored in Amazon S3.

### Why This Step Matters

Launching the EC2 instance completed the cloud environment by combining cloud storage with cloud computing. Amazon S3 stores the MovieLens dataset, while Amazon EC2 provides the virtual computer that could be used to process the data using Apache Spark.

After verifying that the instance launched successfully, I stopped it to avoid unnecessary charges. This demonstrates good cloud resource management while still completing the required deployment.

## Amazon VPC – Network Security

The third AWS service used in this project is Amazon Virtual Private Cloud (Amazon VPC). Amazon VPC provides the secure networking environment that allows AWS resources to communicate with one another. When I launched my Amazon EC2 instance, AWS automatically placed it inside the default VPC and created the necessary network settings for the deployment.

Although I used the default VPC provided by AWS, this experience helped me understand that cloud computing involves more than simply creating virtual computers. The networking component is equally important because it provides a secure environment where cloud resources can communicate safely.

### Why This Step Matters

Amazon S3 stores the data, Amazon EC2 provides the computing resources, and Amazon VPC provides the secure network that allows these services to work together. Even though I did not create a custom VPC for this project, using the default VPC introduced me to how AWS automatically provides secure networking for cloud deployments.

## Advantages and Limitations

### Advantages

Working with AWS showed me how different cloud services can work together to support a data science application. Amazon S3 provided reliable cloud storage for the MovieLens dataset, while Amazon EC2 supplied the computing resources needed to process the data. Using the default Amazon VPC also introduced the networking component that securely connects AWS resources. Another advantage was gaining practical experience with AWS through a Free Tier eligible trial account without purchasing additional computer hardware.

### Limitations

Although AWS provides a powerful cloud platform, there is a learning curve for first-time users because many services and configuration options are available. In addition, while the AWS trial account provides Free Tier eligible resources, cloud services can generate costs once those trial benefits or usage limits are exceeded. Finally, this project demonstrated the deployment process but did not execute the Spark recommender system directly on the EC2 instance.
